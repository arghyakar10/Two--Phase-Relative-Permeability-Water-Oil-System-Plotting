```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sw = np.linspace(0,1,100)
sw
```

```python
so = (1-sw)
so
```

```python
krw=sw**3
krw
```

```python
kro=(1-krw)**2.6
kro
```

```python
df = pd.DataFrame({'Rel Perm Oil':kro,'Rel Perm Water':krw})
df
```

```python
df.plot()
```

```python
#plotting kro and krw vs sw plot
#Labelling and legend to the plots

plt.plot(sw,kro,label='Rel Oil Perm')
plt.plot(sw,krw,label='Rel Water Perm')
plt.legend()
plt.grid()
plt.xlabel('Water Saturation')
plt.ylabel(['Rel Perm to Oil','Rel Perm to Water'])
```

```python
# Here are all cosmetics changes done to made it look beautiful

plt.style.use('classic')
plt.figure(figsize=(12,8))
plt.plot(sw,kro,label='Rel Oil Perm',color='Green',lw=3)
plt.plot(sw,krw,label='Rel Water Perm',color='Blue',lw=3)
plt.legend(loc='upper center')
plt.grid(True)
plt.xlabel('Water Saturation(Sw)')
plt.ylabel(['Rel Perm to Oil (Kro)','Rel Perm to Water (Krw)'],size=16)
plt.axvline(0.2,ls='--',color='Black')
plt.axvline(0.95,ls='--',color='Black')
plt.title('Two- Phase Relative Permeability (Water-Oil System)',size=20)
```
